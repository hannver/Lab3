```python
#------------------------------------------------------------
#------------------------------------------------------------
def NumberToText(input):
   """ 
   Convert number from the range [1..12] to their verbal form
   
   Parameters
   ----------
   input: num
     number
     
   Return
   ------
   str
     expressing numbers in words
   """
   return number_converter.get(input, None)
#------------------------------------------------------------
#------------------------------------------------------------

def pluralize_gift(gift):
  """
  Returns plural of a noun
  
  Parameters
  ----------
  gift: str
    A noun
    
  Return
  ------
  str
    Plural version
  """

  if gift=='goose':
    gift = 'geese'
  elif gift[len(gift)-1]=='y':
       gift=gift.replace('y', 'ies')
  else:
    gift =gift+'s'
  return gift
#-------------------------------------------------------------
#------------------------------------------------------------ 
def make_phrase(num, item, verb, adjective, location):
  """
  Returns phrase of song which is built from input words
  
  Parameters
  ----------
  num:int64
    Number of day
  item:str
    A noun
  verb:str
    A verb 
  adjective:str
    An adjective  
  location:str
     A location    
    
  Return
  ------
  str
    Phrase wich consists from the input words
  """
  first=item[0]
  if (num==1) and ((first=='a')or(first=='e')or(first=='i')or(first=='o')or(first=='u')or(first=='y')):
     item='an '+item
  elif (num==1):
     item='a '+item  
  if num>1:
     item=pluralize_gift(item)
  if num==1:   
     phrase=" ".join([adjective, item, verb, location])
  else:
     phrase=" ".join([NumberToText(num), adjective, item, verb, location])
  phrase=' '.join(phrase.split()) #extra spaces removing
  return phrase
#----------------------------------------------------------------
#----------------------------------------------------------------
def sing_day(data_set, day, col_name):
  """
  Returns verse of song for one day
  
  Parameters
  ----------
  data_set: pandas dataset
    contains information for song
  day: int64
    day number
  col_name: str
     name of the column with phrases 
    
  Return
  ------
  str
    contains a verse for one day
  """
  if day>12 or day<1 :
      return 'no verse for this day'
  # Step 1: Setup the intro line
  num_word = data_set.at[day-1,'Day.in.Words'] # convert "1" to "first" etc.
  intro = "On the " + num_word + " day of Christmas,\nmy true love sent to me:\n"
  # Step 2: Sing the gift phrases
  gifts = ""
  temp=day-1
  for i in range(temp,0,-1):
     gifts = gifts + data_set.at[i,col_name]+',\n'
  # Step 3: Put it all together and return   
  if day==1:
     day_verse=intro+gifts+data_set.at[0,col_name]+'.'
  else:
     day_verse=intro+gifts+'and '+data_set.at[0,col_name]+'.'  
  return day_verse
#--------------------------------------------------------------
#--------------------------------------------------------------
 
def sing_song(data_set, verse):
    """
    Output the specified number of verses of a song
    
    Parameters
    ----------
    data_set: pandas dataset
      contains information about song
    verse: int64
      number of verse to be output
     
      
    Return
    ------
    No values
    """

    for i in range(0,verse,1):
        data_set.at[i,'Phrase']=make_phrase(data_set.iloc[i]['Day'],
                                            data_set.iloc[i]['Gift.Item'], 
                                            data_set.iloc[i]['Verb'],
                                            data_set.iloc[i]['Adjective'],
                                            data_set.iloc[i]['Location'],
                                           )
    for i in range(1,verse+1,1):
        tmp=sing_day(data_set, i,'Phrase')
        print(tmp,'\n') 
             
#---------main code-----------------------------------------
#-------------------------------------------------------------
import pandas as pd 
number_converter = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five",
    6: "six",
    7: "seven",
    8: "eight",
    9: "nine",
    10: "ten",
    11: "eleven",
    12: "twelve"
}

xmas = pd.read_csv("https://www.dropbox.com/scl/fi/qxaslqqp5p08i1650rpc4/xmas.csv?rlkey=erdxi7jbh7pqf9fh4lv4cayp5&dl=1")
xmas.fillna('',inplace=True) 

print("-----The Twelve Days of Christmas------")           
sing_song(xmas,12)           
```
